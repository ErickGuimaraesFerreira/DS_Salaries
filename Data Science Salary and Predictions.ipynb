```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split

# Transformar os dados usando o LabelEncoder
cols = ['experience_level', 'employment_type', 'job_title','salary_currency','employee_residence','company_location','company_size']
df[cols]=df[cols].apply(LabelEncoder().fit_transform)

# Criar X, y
X = df.drop(["salary_in_usd"], axis=1)
y = df["salary_in_usd"]

# Construir conjuntos de treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
import tensorflow as tf
```

```python
tf.random.set_seed(42)

# Criar um modelo
modelo = tf.keras.Sequential([
        tf.keras.layers.Dense(100, activation='relu'),
        tf.keras.layers.Dense(10, activation='relu'),
        tf.keras.layers.Dense(1),
])

# Compilar o modelo
modelo.compile(loss=tf.keras.losses.mae,
              optimizer=tf.keras.optimizers.Adam(),
              metrics=["mae"])

# Adaptar o modelo
histÃ³ria = modelo.fit(X_train, y_train, epochs=150)
```

```python
prediction = modelo.predict(X_test)
```

```python
prediction = np.ravel(prediction)
```

```python
y_pred = modelo.predict(X_test)

plt.figure(figsize=(14, 6))
plt.plot(y_test, label='Valores Reais', marker='o')
plt.plot(y_pred, label='PrevisÃ£o', marker='x')
plt.title('Valores Reais vs PrevisÃ£o')
plt.xlabel('Sample')
plt.ylabel('Valor')
plt.legend()
plt.show()
```

```python
df_final = pd.DataFrame({"Y_test": y_test , "Prediction" : prediction})

# Sort index before plot
df_final = df_final.sort_index()

# Plot the final result
plt.figure(figsize= (18,5))
plt.plot(df_final)
plt.legend(["Atual" , "PrevisÃ£o"])
```
